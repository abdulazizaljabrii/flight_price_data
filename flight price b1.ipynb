# 3. Exploratory Data Analysis :



```python
sns.histplot(data=df, x='source_city', hue='class', multiple='stack')


plt.xlabel('Source City')
plt.ylabel('Count')
plt.title('Histogram of Source City by Class')


plt.show()
```

```python
airline_counts = df['airline'].value_counts()

sns.barplot(x=airline_counts.index, y=airline_counts.values, color='steelblue')

plt.xticks(rotation=90)
plt.xlabel('Airline')
plt.ylabel('Count')
plt.title('Most Frequently Used Airlines')
plt.show()
```

```python
departure_time = df['departure_time'].value_counts()

sns.barplot(x=departure_time.index, y=departure_time.values, color='steelblue')

plt.xticks(rotation=90)
plt.xlabel('Departure Time')
plt.ylabel('Count')
plt.title('Most Common Departure Times')
plt.show()
```

```python
sns.countplot(data=df, x='source_city', color='steelblue')

plt.xticks(rotation=45)
plt.xlabel('Source City')
plt.ylabel('Count')
plt.title('Count Plot of Source Cities')

plt.show()
```

```python
arrival_cities = df['destination_city'].value_counts()


sorted_arrival_cities = arrival_cities.sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='destination_city', order=sorted_arrival_cities.index, color='steelblue')

plt.xticks(rotation=45)
plt.xlabel('Arrival City')
plt.ylabel('Count')
plt.title('Count of Arrival Cities')

plt.show()
```

```python
avg_price_by_city = df.groupby('destination_city')['price'].mean().reset_index()

# Sort the DataFrame based on the 'price' column in descending order
avg_price_by_city = avg_price_by_city.sort_values('price', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(data=avg_price_by_city, x='destination_city', y='price', color='steelblue')

plt.xticks(rotation=45)
plt.xlabel('Destination City')
plt.ylabel('Average Price')
plt.title('Average Price by Destination City')

plt.show()
```

```python

economy_df = df[df['class'] == 'Economy']

avg_price_by_stop = economy_df.groupby('stops')['price'].mean().reset_index()

# Sort the DataFrame based on the 'price' column in descending order
avg_price_by_stop = avg_price_by_stop.sort_values('price', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(data=avg_price_by_stop, x='stops', y='price', color='steelblue')

plt.xticks(rotation=45)
plt.xlabel('Stops')
plt.ylabel('Average Price')
plt.title('Average Price by Stops (Economy Class)')

plt.show()
```

```python

Business_df = df[df['class'] == 'Business']

avg_price_by_stop = Business_df.groupby('stops')['price'].mean().reset_index()

# Sort the DataFrame based on the 'price' column in descending order
avg_price_by_stop = avg_price_by_stop.sort_values('price', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(data=avg_price_by_stop, x='stops', y='price', color='steelblue')

plt.xticks(rotation=45)
plt.xlabel('Stops')
plt.ylabel('Average Price')
plt.title('Average Price by Stops (Business Class)')

plt.show()
```

```python
avg_price_by_city = df.groupby('airline')['price'].mean().reset_index()

avg_price_by_city = avg_price_by_city.sort_values('price', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(data=avg_price_by_city, x='airline', y='price', color='steelblue')

plt.xticks(rotation=45)
plt.xlabel('Airline')
plt.ylabel('Average Price')
plt.title('Average Price by Airline')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='duration', y='price', color='steelblue')

plt.xlabel('Duration')
plt.ylabel('Price')
plt.title('Price vs Duration')

plt.show()
```
